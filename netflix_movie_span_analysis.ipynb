**INVESTIGATING NETFLIX MOVIES LENGTH**

**USE CASE:**
We first define our hypothesis that the average duration of movies has been declining from the initial research. Next we'll delve into the Netflix data to see if we can determine whether movie lengths are actually getting shorter and explain some of the contributing factors, if any.

Now using following dataset named : netflixdata.csv, lets assess EDA skill using the following table detailing the column names and descriptions:

*   The data : netflix_data.csv
*   show_id : The ID of the show
*   type :	Type of show
*   title :	Title of the show
*   director :	Director of the show
*   cast :	Cast of the show
*   country :	Country of origin
*   date_added :	Date added to Netflix
*   release_year :	Year of Netflix release
*   duration :	Duration of the show in minutes
*   duration :	Duration of the show in minutes
*   genre :	Show genre



```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/netflixdata.csv")
print(df.tail())
```

**Subset the DataFrame for "Movies"**

```python
netflix_subset = df[df["type"]== "Movie"]
print(netflix_subset.head())
```

**Subset the columns of the new DataFrame**

```python
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]
print(netflix_movies.tail())
```

**Filter the DataFrame by Movie duration**

```python
short_movies = netflix_movies[netflix_movies["duration"]<60]
print(short_movies.head(20))
```

**Assign colors to movie genres**

```python
colors = []

for index, row in netflix_movies.iterrows():
    genre = row["genre"]

    if genre == "Children":
        colors.append("blue")
    elif genre == "Documentaries":
        colors.append("green")
    elif genre == "Stand-Up":
        colors.append("yellow")
    else:
        colors.append("red")

fig = plt.figure()
plt.scatter(netflix_movies["release_year"], netflix_movies["duration"], c=colors)
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
plt.show()
```
